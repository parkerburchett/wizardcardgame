```python
import numpy as np
import pandas as pd
import itertools
import random
from simulate_wizard import Deck
```

```python
return_deck = lambda : Deck._create_new_deck(include_special=False)
```

the goal is to get the distribuions of counts under code golf (less characters
)

```python
sizes = []

for _ in range (20):
    # you can play 20 shoes in a day est 

    deck = [c for c in return_deck() for _ in range(6)]
    random.shuffle(deck)
    count = []
    for card in deck[150:]:
        if card.number <= 6:
            count.append(1)
        elif (card.number >6) and (card.number < 10):
            count.append(0)
        else:
            count.append(-1)
    count = np.cumsum(count)

    run_size = 0
    max_run_size = 0
    for num in count:
        if num >= 2:
            run_size +=1
            if run_size > max_run_size:
                max_run_size = run_size
        else:
            run_size = 0

    sizes.append(max_run_size)

winners = [x > 50 for x in sizes]
percent = sum(winners) / len(sizes)
```
